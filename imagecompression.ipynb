```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from skimage import io
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Replace with the actual path to your image file in Google Drive
image_path = '/content/drive/My Drive/Colab Notebooks/WhatsApp Image 2025-07-30 at 10.52.01 AM.jpeg'
image = io.imread(image_path)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# Reshape image into (num_pixels, 3)
pixels = image.reshape(-1, 3)

```

```python
kmeans = KMeans(n_clusters=16, random_state=42, n_init=5)
kmeans.fit(pixels)
```

```python
new_colors = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = new_colors.reshape(image.shape).astype(np.uint8)
```

```python
# Plot original vs compressed
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")


plt.subplot(1,2,2)
# Define k from the number of clusters used in KMeans
k = kmeans.n_clusters
plt.imshow(compressed_image)
plt.title(f"Compressed Image (K={k})")


plt.show()
```

```python
# Original image unique colors
orig_colors = np.unique(image.reshape(-1, 3), axis=0)
print("Number of unique colors in original image:", len(orig_colors))

# Compressed image unique colors
comp_colors = np.unique(compressed_image.reshape(-1, 3), axis=0)
print("Number of unique colors in compressed image:", len(comp_colors))
```
